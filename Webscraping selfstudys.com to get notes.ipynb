```python
# Use this Program to get PDF link or download multiple any PDF from selfstudys.com

import requests
from bs4 import BeautifulSoup

# ************PDF downloader***********************
def download_pdf(url, file_name, headers):

    # Send GET request
    response = requests.get(url, headers=headers)

    # Save the PDF
    if response.status_code == 200:
        with open(file_name, "wb") as f:
            f.write(response.content)
    else:
        print(response.status_code)
# *************************************************


# ***********Webscraping Section*******************

print("note = subject link from https://www.selfstudys.com/books/ncert-notes/english/class-9th")
r = requests.get(input("enter link = "))
download_permit = input("Do you Want to Download Pdf ( yes / no )")
 
# Parsing the HTML
soup = BeautifulSoup(r.content, 'html.parser')
 
# Finding by id
s = soup.find('div', class_= 'sample-paper')

for link in s.find_all('a'):

    tex = link.text
    print(tex)
    
    if link.get('class')== ['p-3', 'py-4', 'mb-0']:
        
        l1 = "https://www.selfstudys.com"+link.get('href')
#         print(l1)
        
        r1 = requests.get(l1)
 
        soup1 = BeautifulSoup(r1.content, 'html.parser')
 
        # Finding by id
        s1 = soup1.find('div', id= 'PDFF')
        pdfl = s1.get('source')
        print(pdfl)
        
# *************************************************        
        
        
# ************Downloading section******************      
        
        if download_permit == 'yes':
            
            if __name__ == "__main__":

            # Define HTTP Headers
                headers = {
                    "User-Agent": "Chrome/51.0.2704.103",
                }

                # Define URL of a PDF
                url = pdfl

                # Define PDF file name
                file_name = tex+".pdf"

                # Download PDF
                download_pdf(url, file_name, headers)
    
# *************************************************                

```
